```python
import pandas as pd
```

```python
games = pd.read_csv("games.csv")
```

```python
games.head()
```

```python
games["Team"] = games["Team"].str.replace(r"[\[\]']", "",regex=True)
games["Genres"] = games["Genres"].str.replace(r"[\[\]']", "",regex=True)
games["Reviews"] = games["Reviews"].str.replace(r"[\[\]']", "",regex=True)
```

```python
games.head()
```

```python
games = games.drop(columns=["Unnamed: 0"])
```

```python
games.head()
```

```python
games.isnull().sum()
```

```python
games = games.dropna()
games.head()
```

```python
games = games.drop_duplicates()
games.head()
```

```python
def convert_k(x):
    x = str(x).upper().strip()
    if "K" in x:
        return float(x.replace("K", "")) * 1000
    else:
        return float(x)


games["Times Listed"] = games["Times Listed"].apply(convert_k)
games["Number of Reviews"] = games["Number of Reviews"].apply(convert_k)
games["Plays"] = games["Plays"].apply(convert_k)
games["Playing"] = games["Playing"].apply(convert_k)
games["Backlogs"] = games["Backlogs"].apply(convert_k)
games["Wishlist"] = games["Wishlist"].apply(convert_k)
```

```python
games.head()
```
